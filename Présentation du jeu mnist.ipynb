```python
from tensorflow.examples.tutorials.mnist import input_data
import matplotlib.pyplot as plt
import numpy as np
```

# Présentation du jeu de Données MNIST

Le jeu de données MNIST regroupe 65 000 images de chiffres de 0 à 9 écrits à la main répartis dans 2 dataset, le premier (train) comporte 55 000 images et le second (test) en comporte 10 000. Ces images sont de dimensions 28x28 donc composés de 784 pixels chacunes et sont labélisé. Les classes sont répartis de façon relativement équilibrés.

## Extraction du jeu de données

```python
mnist = input_data.read_data_sets('MNIST_data', one_hot=True)

labels_train = np.dot(mnist.train.labels,np.array(range(10)))
labels_test = np.dot(mnist.train.labels,np.array(range(10)))
```

### LABELS REPARTION

```python

plt.subplot(1,2,1)

plt.hist(labels_train,width = 0.8)
plt.title("Train set label frequency")
plt.ylabel("number")
plt.xlabel("label")

plt.subplot(1,2,2)

plt.hist(labels_test,width = 0.8)
plt.title("Test set label frequency")
plt.ylabel("number")
plt.xlabel("label")

plt.figure(figsize = 10)

plt.show()
```

```python
images = mnist.train.images
```
