```python
from  matplotlib import pyplot as plt
import pandas as pd

%matplotlib inline

plt.style.use('seaborn')

#plt.style.available
```

# Présentation du jeu de données jaffe

Jaffe est un jeu de données de 213 images de dimensions 26x26, et au total de 676 pixels représentant des images de 7 expressions faciales. Ce jeu de données est composé de 10 classes corespondantes à 10 modèles féminin japonais.

# Load data from Jaffe

```python
jaffe_X = pd.read_csv("jaffeX.csv",delimiter=";").values

jaffe_y = pd.read_csv("jaffeY.csv",delimiter=";").values[0]
```

```python
jaffe_X.shape
```

```python
# Affichage de la fréquence de chacun des lables

plt.hist(jaffe_y,width = 0.8, bins=10)

plt.title("Répartition des classes")
plt.ylabel("nombre")
plt.xlabel("classe")

plt.show()
```

## Affichage d'images : 

```python
images = jaffe_X

plt.figure(figsize= (10,10))

for i in range(9):
    
    plt.subplot(3,3,1+i)
    img = images[i].reshape([26,26])
    plt.imshow(img.T,cmap = "gray")

plt.show()
```
