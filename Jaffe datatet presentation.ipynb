```python
from  matplotlib import pyplot as plt
import pandas as pd
```

# Présentation du jeu de données jaffe

Jaffe est un jeu de données de 213 images de dimensions 26x26, et lau total de 676 pixels représentant des images d'expressions faciales. Ce jeu de données est composé de 10 classes corespondantes à 10 émotions différentes.

# Load data from Jaffe

```python
jaffe_X = pd.read_csv("jaffeX.csv",delimiter=";").values

jaffe_y = pd.read_csv("jaffeY.csv",delimiter=";").values[0]
```

```python
jaffe_X.shape
```

```python
# Affichage de la fréquence de chacun des lables

plt.hist(jaffe_y,width = 0.8)

plt.title("Train set label frequency")
plt.ylabel("number")
plt.xlabel("label")

plt.show()
```

## Affichage d'images : 

```python
images = jaffe_X

plt.figure(figsize= (10,10))

for i in range(9):
    
    plt.subplot(3,3,1+i)
    img = images[i].reshape([26,26])
    plt.imshow(img.T,cmap = "gray")
    
    
plt.show()
```
